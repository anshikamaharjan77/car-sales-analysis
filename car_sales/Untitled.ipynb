```python
import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("car_data.csv")
```

```python
df.head()
```

WHICH MANUFACTURE HAS THE MOST CAR

```python
manufacturer_counts=df['Manufacturer'].value_counts()
```

```python
plt.figure(figsize=(6,3))
plt.bar(manufacturer_counts.index, manufacturer_counts.values, color='skyblue')
plt.title('Number of Cars by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

```python
#AVERAGE PRICE FOR EACH MANUFACTURER
avg_price=df.groupby('Manufacturer')['Price'].mean()
```

```python
plt.figure(figsize=(6,3))
plt.bar(avg_price.index,avg_price.values,color='orange')
plt.title("Average Price")
plt.xlabel('Manufacturer')
plt.ylabel('Average Price')
plt.show()
```

```python
#Relationship between engine size and price
plt.figure(figsize=(6,6))
plt.scatter(df['Engine size'],df['Price'],color='red')
plt.title('Relationship between Engine Size and Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

```python
#Count of each fuel type
fuel_counts=df['Fuel type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(fuel_counts,labels=fuel_counts.index,autopct='%0.1f%%')
plt.title('Count of each Fuel type')
plt.show()
```

```python
#Number of cars from each year
year_counts=df['Year of manufacture'].value_counts()
plt.figure(figsize=(6,6))
plt.plot(year_counts.index,year_counts.values,marker='o')
plt.title('Number of Cars by Year')
plt.xlabel('Year of Manufacturer')
plt.ylabel('Number of Cars')
plt.show()
```

```python
#Price distribution by Manufacturer
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Manufacturer', y='Price')
plt.title('Price Distribution by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Price')
plt.show()
```

```python
#What is the median price for each manufacturer.
med_price=df.groupby('Manufacturer') ['Price'].median()
plt.figure(figsize=(6,6))
plt.bar(med_price.index,med_price.values,color='pink')
plt.title('Median price of each manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Compare price distribution by fuel type
plt.figure(figsize=(6,6))
sns.boxplot(data=df,x='Fuel type',y='Price')
plt.title('Price distribution by fuel type')
plt.xlabel('Fuel type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#What is the highest price of fuel type
hig=df.groupby('Fuel type') ['Price'].max()
print(hig)
```

```python
#What is the lowest price of fuel type
low=df.groupby('Fuel type') ['Price'].min()
print(low)
```

```python
#Distribution of all car price
plt.figure(figsize=(6,6))
plt.hist(df['Price'], bins=30 , color='skyblue', edgecolor='black')
plt.title('Distribution of all car price')
plt.xlabel('Price')
plt.ylabel('Number of cars')
plt.show()
```

Mileage and Engine Size Analysis

```python
#Average Mileage per Manufacturer
avg_mileage=df.groupby('Manufacturer') ['Mileage'].mean()
plt.figure(figsize=(3,3))
plt.bar(avg_mileage.index, avg_mileage.values, color='green')
plt.title('Average Mileage per Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Mileage')
plt.xticks(rotation=45)
plt.show()
```

```python
#Mileage VS Price
plt.figure(figsize=(10,10))
plt.scatter(df['Mileage'],df['Price'],color='pink', s=50, alpha=0.6)
plt.grid(True)
plt.title('Mileage VS Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
#Engine Size Distribution
plt.figure(figsize=(5,5))
plt.hist(df['Engine size'], bins=30,color='red',edgecolor='black')
plt.title('Engine size distribution')
plt.xlabel('Engine Size')
plt.ylabel('Number of Cars')
plt.show()
```

Advanced visualization

```python
#Pair Plot of Pirce, Mileage and Engine Size
df_pair = df[['Price', 'Mileage', 'Engine size']]
sns.pairplot(df_pair)
plt.show()
```

```python
#Heatmap of average price per manufacturer vs fuel type
pivot_table=df.pivot_table(values='Price',index='Manufacturer',columns='Fuel type', aggfunc='mean')
plt.figure(figsize=(6,6))
sns.heatmap(pivot_table,annot=True, fmt=".0f",cmap="coolwarm")
plt.title('Heatmap of average price : manufacturer vs fuel type')
plt.show()
```

```python
#Boxplot of mileage per fuel type per manufacturer
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='Manufacturer',y='Mileage',hue='Fuel type')
plt.title('Mileage per Fuel type per Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Mileage')
plt.xticks(rotation=45) 
plt.legend(title='Fuel type')
plt.show()
```

#Dashboard(2X2 Subplots Example

```python
fig,axs=plt.subplots(2,2,figsize=(12,10))

#1. Histogram of Price
axs[0,0].hist(df['Price'],bins=20,color='skyblue', edgecolor='black')
axs[0,0].set_title('Distribution of Car Prices')
axs[0,0].set_xlabel('Price')
axs[0,0].set_ylabel('Count')

#2. Mileage vs Price Scatter
axs[0,1].scatter(df['Mileage'],df['Price'],color='pink')
axs[0,1].set_title('Mileage vs Price ')
axs[0,1].set_xlabel('Mileage')
axs[0,1].set_ylabel('Price')

#3. Engine Size vs Price Scatter
axs[1,0].scatter(df['Engine size'],df['Price'],color='red')
axs[1,0].set_title('Engine size vs Price ')
axs[1,0].set_xlabel('Engine Size')
axs[1,0].set_ylabel('Price')

#4. Barchart of Average Mileage per Manufacturer
avg_mil=df.groupby('Manufacturer')['Mileage'].mean()
axs[1,1].bar(avg_mil.index, avg_mil.values, color='Brown')
axs[1,1].set_title('Average Mileage per Manufacturer')
axs[1,1].set_xlabel('Manufacturer')
axs[1,1].set_ylabel('Mileage')
axs[1,1].tick_params(axis='x',rotation=45)
plt.tight_layout()
plt.show()
```
